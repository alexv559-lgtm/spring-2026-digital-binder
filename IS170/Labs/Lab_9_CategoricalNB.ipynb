```python
#import library needed to run categorical NB
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
```

```python
#Load the dataset
df = pd.read_csv("food_preference.csv")
```

```python
#Display the data set to see what columns we are working with
df
```

```python
#Dropped Timestamp, participant id, and age since they aren't needed for the prediction
df.drop(["Timestamp","Participant_ID","Age"], axis=1, inplace=True)
df
```

```python
#Check for missing values in each column
print(df.isnull().sum())
```

```python
#Drop the missing values
#Display the cleaned dataset
df.dropna(inplace=True)
df
```

```python
#Verify all missing values are removed
print(df.isnull().sum())
```

```python
#Create LabelEncoders to convert categorical values into numbers
#Each column needs its own encoder
encoder_gender = LabelEncoder()
encoder_nationality = LabelEncoder()
encoder_food = LabelEncoder()
encoder_juice = LabelEncoder()
encoder_target = LabelEncoder()

#Convert each column from words to numbers
df["Gender_num"] = encoder_gender.fit_transform(df["Gender"])
df["Nationality_num"] = encoder_nationality.fit_transform(df["Nationality"])
df["Food_num"] = encoder_food.fit_transform(df["Food"])
df["Juice_num"] = encoder_juice.fit_transform(df["Juice"])
df["Dessert_num"] = encoder_target.fit_transform(df["Dessert"])

print(df[["Gender", "Gender_num", "Nationality","Nationality_num","Food", "Food_num","Juice", "Juice_num","Dessert","Dessert_num"
]])
print()
```

```python
#Define features (X) and target (y)
#X is used to make predictions
#y what we are trying to predict
X = df[["Gender_num", "Nationality_num", "Food_num", "Juice_num"]]
y = df["Dessert_num"]
```

```python
#Create and train the Categorical Naive Bayes model
model = CategoricalNB()
model.fit(X, y)
```

```python
#Here we create a new sample to test the model
#This is Data the model has never seen before
new_data = [["Male", "Indian", "Western Food", "Fresh Juice"]]

#Convert the new data using the same encoders used for training
#Here we use transform so we don't relearn categories
new_input = [[
    encoder_gender.transform([new_data[0][0]])[0],
    encoder_nationality.transform([new_data[0][1]])[0],
    encoder_food.transform([new_data[0][2]])[0],
    encoder_juice.transform([new_data[0][3]])[0]
]]

#Make prediction using the trained model
prediction = model.predict(new_input)[0]

#Get the probability of each possible outcome
probabilities = model.predict_proba(new_input)[0]

#Convert numeric prediction back to original label
result = encoder_target.inverse_transform([prediction])[0]
```

```python
#Here we display the prediction which was a Yes to dessert.
print("Prediction:", result)
print("Porbabilities:", probabilities)
```

The categoricalNB model uses categorical data like gender, nationality, food, and juice to predict a person's dessert preference. The dataset was cleaned by removing unnecessary columns and dropping rows with missing values to make sure the model runs properly. Then we converted the categorical values into numbers using labelencoder so the model can process them. After that the data was split into features (X) and the target (y), and a categorical naive bayes model was trained. Lastly we added a new sample to test the model and it returned a prediction and probabilities for each outcome. This model shows using probability to make a decision based on categorical inputs.
