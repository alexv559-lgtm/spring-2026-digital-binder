```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('cybersecurity_intrusion_data.csv')
df = df.replace('None', np.nan)
df = df.replace('Unknown', np.nan)
df.isnull().sum()
```

```python
df.dropna(subset=['encryption_used'], inplace=True)
df.dropna(subset=['protocol_type'], inplace=True)
df.dropna(subset=['browser_type'], inplace=True)
df = df.drop(columns= ['session_id'], axis=1)
df.sample(10)
```

```python
le = LabelEncoder()

label = le.fit_transform(df['protocol_type'])
label = le.fit_transform(df['encryption_used'])
label = le.fit_transform(df['browser_type'])
label

df.drop('protocol_type', axis=1, inplace=True)
df['protocol_type'] = label
df.drop('encryption_used', axis=1, inplace=True)
df['encryption_used'] = label
df.drop('browser_type', axis=1, inplace=True)
df['browser_type'] = label
df
```

```python
#Save columns into cols to keep columns for standard scaler
cols = [col for col in df.columns if col != 'attack_detected']
#Stand scaler standardizes the data, while keeping the index of the original columns
ss = StandardScaler()
df[cols] = ss.fit_transform(df[cols])
df
```

```python
#split dataset in features and target variable
x_data=df[['network_packet_size',	'login_attempts',	'session_duration',	'ip_reputation_score',	'failed_logins',	'unusual_time_access',	'protocol_type',	'encryption_used',	'browser_type']]
y_data=df['attack_detected']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)
```

```python
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train,y_train)

# Calculate the accuracy and assign it to dt_accuracy
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Calculate the accuracy and assign it to dt_accuracy
dt_accuracy = dt_model.score(X_test, y_test)

print("Decision Tree Test Accuracy:", f"{dt_accuracy * 100:.2f}%")
```

```python
# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)


# Calculate the accuracy and assign it to rf_accuracy
rf_accuracy = rf_model.score(X_test, y_test)


# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split


rf_model.fit(X_train, y_train)


# Calculate the accuracy and assign it to rf_accuracy
rf_accuracy = rf_model.score(X_test, y_test)

# Now you can use rf_accuracy in the print statements below
print("RF Model Accuracy:", f"{rf_accuracy * 100:.2f}", "%")

```

```python

# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate precision, recall, and F1-score for Random Forest
precision_rf = metrics.precision_score(y_test, y_pred_rf)
recall_rf = metrics.recall_score(y_test, y_pred_rf)
# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate precision, recall, and F1-score for Random Forest
precision_rf = metrics.precision_score(y_test, y_pred_rf)
recall_rf = metrics.recall_score(y_test, y_pred_rf)
f1_rf = metrics.f1_score(y_test, y_pred_rf)


print("RF Precision:", precision_rf)
print("RF Recall:", recall_rf)
print("RF F1-Score:", f1_rf)
```

```python
# Prepare a provision for a new data entry and its prediction of an outcome.

def predict_intrusion():
    while True:
        try:
            packet_size = float(input("Enter packet size: "))
            protocol_type = float(input("Enter protocol type: "))
            encryption_used = float(input("Enter encryption used: "))
            browser_type = float(input("Enter browser type: "))
            login_attempts = float(input("Enter login attempts: "))
            session_duration = float(input("Enter session duration: "))
            ip_reputation_score = float(input("Enter IP reputation score: "))
            failed_logins = float(input("Enter failed logins: "))
            unusual_time = float(input("Enter unusual time access: "))


            # Input validation (example: check for non-negative values)
            if any(val < 0 for val in [packet_size, protocol_type, encryption_used, browser_type, login_attempts, session_duration, ip_reputation_score, failed_logins, unusual_time]):
                print("Error: Please enter non-negative values for all features.")
                continue  #restart the loop

            new_intrusion_data = pd.DataFrame([[packet_size, protocol_type, encryption_used, browser_type, login_attempts, session_duration, ip_reputation_score, failed_logins, unusual_time]],
                                           columns=['network_packet_size',	'login_attempts',	'session_duration',	'ip_reputation_score',	'failed_logins',	'unusual_time_access',	'protocol_type',	'encryption_used',	'browser_type'])

            prediction =rf_model.predict(new_intrusion_data)[0]

            if prediction == 1:
                print("Prediction: Likely a intrusion.")
            else:
                print("Prediction: Not likely a intrusion.")

            another_intrusion = input("Enter data for another intrusion? (yes/no): ")
            if another_intrusion.lower() != 'yes':
                break #exit loop
        except ValueError:
            print("Invalid input. Please enter valid numerical values.")

predict_intrusion()
```
