```python
df = df.drop(columns=['Fault_Type'], axis=1)
df
```

```python
#split dataset in features and target variable
x_data=df[['S11 (dB)',	'VSWR',	'Gain (dBi)',	'Eff (%)',	'BW (MHz)',	'Z_Real (Ohms)',	'Z_Imag (Ohms)']]
y_data=df['Status']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)
```

```python
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train,y_train)

# Calculate the accuracy and assign it to dt_accuracy
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Calculate the accuracy and assign it to dt_accuracy
dt_accuracy = dt_model.score(X_test, y_test)

print("Decision Tree Test Accuracy:", f"{dt_accuracy * 100:.2f}%")
```

```python
# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)


# Calculate the accuracy and assign it to rf_accuracy
rf_accuracy = rf_model.score(X_test, y_test)


# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split


rf_model.fit(X_train, y_train)


# Calculate the accuracy and assign it to rf_accuracy
rf_accuracy = rf_model.score(X_test, y_test)

# Now you can use rf_accuracy in the print statements below
print("RF Model Accuracy:", f"{rf_accuracy * 100:.2f}", "%")
```

```python
# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate precision, recall, and F1-score for Random Forest
precision_rf = metrics.precision_score(y_test, y_pred_rf)
recall_rf = metrics.recall_score(y_test, y_pred_rf)
# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate precision, recall, and F1-score for Random Forest
precision_rf = metrics.precision_score(y_test, y_pred_rf)
recall_rf = metrics.recall_score(y_test, y_pred_rf)
f1_rf = metrics.f1_score(y_test, y_pred_rf)


print("RF Precision:", precision_rf)
print("RF Recall:", recall_rf)
print("RF F1-Score:", f1_rf)
```

```python
def predict_status():
    while True:
        try:
            s11 = int(input('Enter s11 value: '))
            vswr = int(input('Enter vswr value: '))
            gain = int(input('Enter gain value: '))
            eff = int(input('Enter eff'))
            bw = int(input('Enter bw value: '))
            z_real = int(input('Enter z_real value: '))
            z_imag = int(input('Enter z_imag value: '))

            # Input validation (example: check for non-negative values)
            if any(val < 0 for val in [s11, vswr, gain, eff, bw, z_real, z_imag]):
                print("Error: Please enter non-negative values for all features.")
                continue  #restart the loop

            new_antenna_data = pd.DataFrame([[s11, vswr, gain, eff, bw, z_real, z_imag]],
                                           columns=['S11 (dB)',	'VSWR',	'Gain (dBi)',	'Eff (%)',	'BW (MHz)',	'Z_Real (Ohms)',	'Z_Imag (Ohms)'])

            prediction =rf_model.predict(new_antenna_data)[0]

            if prediction == 1:
                print("Prediction: Antenna is likely to be faulty.")
            else:
                print("Prediction: Antenna is likely not to be faulty.")

            another_antenna = input("Enter data for another antenna? (yes/no): ")
            if another_antenna.lower() != 'yes':
                break #exit loop
        except ValueError:
            print("Invalid input. Please enter valid numerical values.")

predict_status()
```
